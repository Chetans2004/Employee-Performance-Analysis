#  Performance Prediction

```python
# ==============================
# PREDICT MODEL 
# ==============================

# Import libraries
import pandas as pd
import joblib

# ------------------------------
# Load Trained Model
# ------------------------------
model = joblib.load('../models/employee_performance_model.pkl')

print( Model loaded successfully!")

# ------------------------------
# Load Test Data
# ------------------------------
X_test = pd.read_csv('../../data/processed/X_test.csv')

print("\nTest Data Loaded:")
print(X_test.head())

# ------------------------------
# Make Predictions
# ------------------------------
predictions = model.predict(X_test)

print("\nSample Predictions:")
print(predictions[:10])

# ------------------------------
# Save Predictions
# ------------------------------
output = pd.DataFrame({
    'Predicted_Performance': predictions
})

output.to_csv('../../data/processed/predictions.csv', index=False)

print("\n Predictions saved to predictions.csv")

# ------------------------------
# Predict for Single New Employee
# ------------------------------
sample = X_test.iloc[0:1]

single_prediction = model.predict(sample)

print("\nSingle Employee Prediction:", single_prediction[0])
```
