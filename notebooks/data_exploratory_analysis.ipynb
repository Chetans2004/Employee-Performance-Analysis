## 2. Bivariate Analysis

This section explores relationships between pairs of variables, with a focus on **PerformanceRating** as the target variable and other key features.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='PerformanceRating', y='EmpJobSatisfaction', data=df, palette='Set2')
plt.title('Performance Rating vs. Employee Job Satisfaction')
plt.xlabel('Performance Rating')
plt.ylabel('Job Satisfaction')
plt.show()

# Mean satisfaction by rating
print("Mean Job Satisfaction by Performance Rating:")
print(df.groupby('PerformanceRating')['EmpJobSatisfaction'].mean())
```

**INSIGHT:**
- Employees with higher performance ratings tend to have slightly higher job satisfaction.
- However, the variance within each rating group is large — satisfaction alone is not a reliable predictor of performance.
- Employees rated 4 (Outstanding) show a wider range of satisfaction levels.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='PerformanceRating', y='EmpEnvironmentSatisfaction', data=df, palette='Set3')
plt.title('Performance Rating vs. Environment Satisfaction')
plt.xlabel('Performance Rating')
plt.ylabel('Environment Satisfaction')
plt.show()

print("Mean Environment Satisfaction by Performance Rating:")
print(df.groupby('PerformanceRating')['EmpEnvironmentSatisfaction'].mean())
```

**INSIGHT:**
- Higher-performing employees generally report higher environment satisfaction.
- This aligns with the hypothesis that a positive work environment supports better performance.
- Employees rated 2 (Low) show the lowest median environment satisfaction.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='PerformanceRating', y='EmpWorkLifeBalance', data=df, palette='Pastel1')
plt.title('Performance Rating vs. Work-Life Balance')
plt.xlabel('Performance Rating')
plt.ylabel('Work-Life Balance')
plt.show()

print("Mean Work-Life Balance by Performance Rating:")
print(df.groupby('PerformanceRating')['EmpWorkLifeBalance'].mean())
```

**INSIGHT:**
- Work-life balance shows a moderate positive relationship with performance rating.
- Employees with better work-life balance tend to have slightly higher performance.
- This suggests that organizational support for work-life balance may positively impact productivity.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='PerformanceRating', y='EmpLastSalaryHikePercent', data=df, palette='Blues')
plt.title('Performance Rating vs. Last Salary Hike %')
plt.xlabel('Performance Rating')
plt.ylabel('Salary Hike %')
plt.show()

print("Mean Salary Hike % by Performance Rating:")
print(df.groupby('PerformanceRating')['EmpLastSalaryHikePercent'].mean())
```

**INSIGHT:**
- There is a clear positive relationship between salary hike percentage and performance rating.
- Employees rated 4 receive significantly higher salary increments than those rated 2 or 3.
- This confirms that the compensation system rewards higher performance, which may also incentivize it.

```python
ot_perf = df.groupby(['OverTime', 'PerformanceRating']).size().unstack(fill_value=0)
ot_perf_pct = ot_perf.div(ot_perf.sum(axis=1), axis=0) * 100

ot_perf_pct.plot(kind='bar', stacked=True, colormap='Set2', figsize=(8, 5))
plt.title('Performance Rating Distribution by Overtime Status')
plt.xlabel('OverTime')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Performance Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Performance Rating distribution by OverTime (%):")
print(ot_perf_pct.round(2))
```

**INSIGHT:**
- Employees who work overtime show a slightly different performance distribution compared to those who do not.
- Overtime may be associated with higher work demands, which can either motivate or burnout employees.
- The proportion of high performers (Rating 4) is notable in both groups, suggesting overtime alone is not a defining factor.

```python
dept_perf = df.groupby(['EmpDepartment', 'PerformanceRating']).size().unstack(fill_value=0)
dept_perf_pct = dept_perf.div(dept_perf.sum(axis=1), axis=0) * 100

dept_perf_pct.plot(kind='bar', stacked=True, colormap='Pastel1', figsize=(12, 6))
plt.title('Performance Rating Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Performance Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Performance Rating distribution by Department (%):")
print(dept_perf_pct.round(2))
```

**INSIGHT:**
- Performance distribution varies significantly across departments.
- Some departments (e.g., Research & Development, Sales) have a higher proportion of top performers.
- This could reflect department-specific culture, management quality, or the nature of roles.
- HR and Finance departments may have a higher proportion of average performers.

```python
gender_perf = df.groupby(['Gender', 'PerformanceRating']).size().unstack(fill_value=0)
gender_perf_pct = gender_perf.div(gender_perf.sum(axis=1), axis=0) * 100

gender_perf_pct.plot(kind='bar', stacked=True, colormap='Set1', figsize=(8, 5))
plt.title('Performance Rating Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Performance Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Performance Rating distribution by Gender (%):")
print(gender_perf_pct.round(2))
```

**INSIGHT:**
- Performance rating distribution across genders appears relatively similar.
- Any minor differences should be interpreted cautiously due to potential sample size imbalance.
- This suggests that gender is not a strong differentiating factor for performance ratings in this dataset.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='PerformanceRating', y='TotalWorkExperienceInYears', data=df, palette='muted')
plt.title('Performance Rating vs. Total Work Experience (Years)')
plt.xlabel('Performance Rating')
plt.ylabel('Total Work Experience (Years)')
plt.show()

print("Mean Total Work Experience by Performance Rating:")
print(df.groupby('PerformanceRating')['TotalWorkExperienceInYears'].mean())
```

**INSIGHT:**
- Employees with more total work experience tend to achieve higher performance ratings.
- The violin plot reveals the distribution shape — high performers have a broader experience range.
- Experience brings institutional knowledge and domain expertise that likely contribute to higher ratings.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='EmpJobLevel', hue='PerformanceRating', data=df, palette='Set2')
plt.title('Performance Rating by Job Level')
plt.xlabel('Job Level')
plt.ylabel('Count')
plt.legend(title='Performance Rating')
plt.show()

job_perf = df.groupby(['EmpJobLevel', 'PerformanceRating']).size().unstack(fill_value=0)
print("Performance Rating distribution by Job Level:")
print(job_perf)
```

**INSIGHT:**
- Job level has a visible influence on performance rating distribution.
- Entry-level employees (Level 1) are concentrated at average ratings, while senior levels tend to have a greater spread.
- Higher job levels may have more demanding and visible performance expectations, leading to more differentiated ratings.

```python
num_cols = ['Age', 'DistanceFromHome', 'EmpEducationLevel', 'EmpEnvironmentSatisfaction',
           'EmpHourlyRate', 'EmpJobInvolvement', 'EmpJobLevel', 'EmpJobSatisfaction',
           'NumCompaniesWorked', 'EmpLastSalaryHikePercent', 'EmpRelationshipSatisfaction',
           'TotalWorkExperienceInYears', 'TrainingTimesLastYear', 'EmpWorkLifeBalance',
           'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole',
           'YearsSinceLastPromotion', 'YearsWithCurrManager', 'PerformanceRating']

corr = df[num_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr[['PerformanceRating']].sort_values('PerformanceRating', ascending=False),
            annot=True, fmt='.2f', cmap='RdYlGn', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation of Numerical Features with PerformanceRating')
plt.tight_layout()
plt.show()

print("Top correlations with PerformanceRating:")
print(corr['PerformanceRating'].sort_values(ascending=False))
```

**INSIGHT:**
- **EmpLastSalaryHikePercent** has the highest positive correlation with PerformanceRating — confirming that salary hikes and performance are tightly linked.
- **EmpEnvironmentSatisfaction**, **EmpJobSatisfaction**, and **EmpWorkLifeBalance** show moderate positive correlations.
- **YearsSinceLastPromotion** shows a slight negative correlation — employees who have not been promoted recently may be less motivated or recognized.
- Demographic features like Age and DistanceFromHome show minimal correlation with performance.

## Bivariate Analysis — Summary

**Key Findings:**

1. **Salary Hike %** is the strongest predictor of performance rating, reflecting the company's reward-performance alignment.
2. **Environment Satisfaction**, **Job Satisfaction**, and **Work-Life Balance** are the top engagement factors positively correlated with performance.
3. **Department** plays a significant role — performance distribution varies by department, suggesting culture and management impact.
4. **Work Experience** matters — more experienced employees tend to achieve higher ratings.
5. **OverTime** shows nuanced effects — it does not strongly distinguish performance levels alone.
6. **Gender and Age** are not significant differentiators for performance in this dataset.

**Top 3 factors driving PerformanceRating:**

3. EmpJobSatisfaction


## 3. Multivariate Analysis

This section explores interactions between **three or more variables** to uncover deeper patterns affecting **PerformanceRating**.

### 3.1 Pairplot — Top Correlated Numerical Features

Visualizing pairwise relationships among the most influential numerical features, colored by **PerformanceRating**.

```python
top_features = [
    'EmpLastSalaryHikePercent',
    'EmpEnvironmentSatisfaction',
    'EmpJobSatisfaction',
    'EmpWorkLifeBalance',
    'TotalWorkExperienceInYears',
    'PerformanceRating'
]

sns.pairplot(df[top_features], hue='PerformanceRating', palette='Set1', diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot of Top Features by Performance Rating', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

**INSIGHT:**
- **Salary Hike % vs Performance Rating** shows a clear upward trend — higher hikes align with higher ratings.
- Clusters for Rating 3 dominate most plots due to class imbalance.
- **Job Satisfaction** and **Environment Satisfaction** together help separate Rating 4 employees from the rest.
- **Work Experience** combined with **Satisfaction scores** is a strong differentiator for top performers.

### 3.2 Department × Job Satisfaction → Mean Performance Rating

A pivot heatmap showing how department and job satisfaction jointly influence performance.

```python
pivot = df.pivot_table(
    values='PerformanceRating',
    index='EmpDepartment',
    columns='EmpJobSatisfaction',
    aggfunc='mean'
)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='YlOrRd', linewidths=0.5)
plt.title('Mean Performance Rating: Department × Job Satisfaction')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Department')
plt.tight_layout()
plt.show()
```

**INSIGHT:**
- Performance varies significantly across the Department × Job Satisfaction grid.
- In most departments, **higher job satisfaction (level 4)** corresponds to the highest mean performance ratings.
- Some departments (e.g., Research & Development) maintain relatively consistent performance even at lower satisfaction levels.
- **HR and Finance** departments show the most variation — satisfaction level strongly predicts performance there.

### 3.3 Full Numerical Correlation Matrix

Overall relationships among all numeric variables.

```python
num_cols = ['Age', 'DistanceFromHome', 'EmpEducationLevel', 'EmpEnvironmentSatisfaction',
           'EmpHourlyRate', 'EmpJobInvolvement', 'EmpJobLevel', 'EmpJobSatisfaction',
           'NumCompaniesWorked', 'EmpLastSalaryHikePercent', 'EmpRelationshipSatisfaction',
           'TotalWorkExperienceInYears', 'TrainingTimesLastYear', 'EmpWorkLifeBalance',
           'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole',
           'YearsSinceLastPromotion', 'YearsWithCurrManager', 'PerformanceRating']

corr_matrix = df[num_cols].corr()

plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, vmin=-1, vmax=1, linewidths=0.3, annot_kws={'size': 7})
plt.title('Full Correlation Matrix — Numerical Features', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**INSIGHT:**
- **ExperienceYearsAtThisCompany** is highly correlated with **YearsWithCurrManager** and **ExperienceYearsInCurrentRole** — multicollinearity to handle during modeling.
- **TotalWorkExperienceInYears** and **Age** are naturally correlated.
- **EmpJobLevel** correlates strongly with **TotalWorkExperienceInYears** — senior roles require experience.
- **PerformanceRating** has the strongest correlation with **EmpLastSalaryHikePercent**, confirming our earlier findings.
- Several satisfaction metrics are weakly correlated with each other, suggesting they measure distinct dimensions of employee well-being.

## Multivariate Analysis — Summary

**Key Multivariate Insights:**

1. **Salary Hike % remains the dominant predictor** of performance across all interaction analyses.
2. **Department + Job Satisfaction** jointly predict performance — some departments leverage satisfaction more effectively than others.
3. **Multicollinearity exists** among experience-related features — feature selection will be critical before modeling.

**Features recommended for modeling:**
- `EmpLastSalaryHikePercent`
- `EmpEnvironmentSatisfaction`
- `EmpJobSatisfaction`
- `EmpWorkLifeBalance`
- `EmpDepartment`
- `EmpJobLevel`
- `TotalWorkExperienceInYears`
- `EmpJobInvolvement`
- `OverTime`
